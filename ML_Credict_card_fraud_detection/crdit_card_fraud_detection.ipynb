# **Dealing with Imbalanced Data**

**Under Sampling Method**



```python
legit_count , fraud_count = credit_card_data['Class'].value_counts()
print(legit_count)
print(fraud_count)
```

```python
legit_under_sample = legit.sample(fraud_count)
print(legit_under_sample.shape)
```

```python
legit_under_sample['Class'].value_counts()
```

```python
# comcatenating the two dataframes
new_data = pd.concat([legit_under_sample,fraud],axis=0)
new_data['Class'].value_counts()
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data.groupby('Class').mean()
```

```python
# splitting the data into features and target
X = new_data.drop(columns='Class',axis=1)
Y = new_data['Class']
print(X)
print(Y)
```

```python
# train and test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2, stratify=Y)
print(X.shape, X_train.shape, X_test.shape)
```

```python
# model training
model = LogisticRegression(max_iter=10000)
model.fit(X_train, Y_train)
```

```python
#evaluate the model
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
training_data_accuracy
```

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
testing_data_accuracy
```
