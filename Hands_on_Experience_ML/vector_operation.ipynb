```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()

plt.quiver(0,0,1,1)

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.quiver(0,0,4,5, scale=1, scale_units='xy', angles='xy', color='b')

plt.show()

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.quiver(0,0,4,5, scale=1, scale_units='xy', angles='xy', color='b')
plt.quiver(0,0,6,7, scale=1, scale_units='xy', angles='xy', color='r')

plt.show()
```

```python
#addition of vectors
vector1 = np.array([0,0,2,3])
vector2 = np.array([0,0,3,2])
vector3 = vector1 + vector2
print(vector3)

plt.xlim(-8,8)  
plt.ylim(-8,8)
plt.quiver(0,0,2,3, scale=1, scale_units='xy', angles='xy', color='b')  
plt.quiver(0,0,3,-2, scale=1, scale_units='xy', angles='xy', color='r')  
plt.quiver(0,0,5,1, scale=1, scale_units='xy', angles='xy', color='g')
plt.show()  
```

```python
#subtraction of vectors
vector1 = np.array([0,0,2,3])
vector2 = np.array([0,0,3,2])
vector3 = vector1 - vector2
print(vector3)

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.quiver(0,0,2,3, scale=1, scale_units='xy', angles='xy', color='b')
plt.quiver(0,0,3,-2, scale=1, scale_units='xy', angles='xy', color='r')
plt.quiver(0,0,-1,5, scale=1, scale_units='xy', angles='xy', color='g')
plt.show()
```

```python
#multiplication of vectors by scalar
vector1 = np.array([0,0,2,3])
vector2 = 2 * vector1
print(vector2)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.quiver(0,0,2,3, scale=1, scale_units='xy', angles='xy', color='b')
plt.quiver(0,0,4,6, scale=1, scale_units='xy', angles='xy', color='r')
plt.show()
```

```python
#multiplication of vectors by scalar
vector1 = np.array([0,0,2,3])
vector2 = -1.5 * vector1
print(vector2)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.quiver(0,0,2,3, scale=1, scale_units='xy', angles='xy', color='b')
plt.quiver(0,0,-3,-4.5, scale=1, scale_units='xy', angles='xy', color='r')
plt.show()
```

```python
# cross and dot product and projection of vectors
# vectors are arrays of numbers

# dot product
vector1 = np.array([2,3])
vector2 = np.array([3,2])   
dot_product = np.dot(vector1, vector2)
print(dot_product)
```

```python
# cross product
vector1 = np.array([2,3])
vector2 = np.array([3,2])
cross_product = np.cross(vector1, vector2)
print(cross_product)
```

```python
# projection of vector1 on vector2
vector1 = np.array([2,3])
vector2 = np.array([3,2])   
magnitude = np.sqrt(sum(vector2**2))
projection = (np.dot(vector1, vector2)/magnitude) * vector2
print("projection of vector1 on vector2",projection)
```
